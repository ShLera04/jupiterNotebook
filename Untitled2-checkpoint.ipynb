```python
import gymnasium as gym
import numpy as np
import random
```

```python
env = gym.make('CartPole-v1', render_mode="human")
obs, info = env.reset()

done = False
total_reward = 0

while not done:
    env.render()
    action = env.action_space.sample()
    obs, reward, terminated, truncated, info = env.step(action)
    total_reward += reward
    done = terminated or truncated

print(f"Общая награда: {total_reward}")
env.close()
```

```python
def discretize_state(obs, bins):
    state_idx = []
    for i in range(len(obs)):
        idx = np.digitize(obs[i], bins[i])
        # Ограничиваем индекс, чтобы не выйти за пределы Q-таблицы
        idx = min(idx, len(bins[i]) - 1)
        state_idx.append(idx)
    return tuple(state_idx)
```

```python
# Границы для дискретизации
pos_bins = np.linspace(-2.4, 2.4, 10)
vel_bins = np.linspace(-3, 3, 10)
angle_bins = np.linspace(-0.5, 0.5, 10)
angle_vel_bins = np.linspace(-2, 2, 10)

bins = [pos_bins, vel_bins, angle_bins, angle_vel_bins]
```

```python
def train_cartpole():
    env = gym.make('CartPole-v1')

    # Создаём Q-таблицу: (10, 10, 10, 10) -> 2 действия
    q_table = np.zeros((10, 10, 10, 10, 2))

    alpha = 0.1
    gamma = 0.95
    epsilon = 1.0
    epsilon_decay = 0.995
    epsilon_min = 0.01

    episodes = 10000

    for ep in range(episodes):
        state, _ = env.reset()
        state_d = discretize_state(state, bins)
        done = False

        while not done:
            if random.uniform(0, 1) < epsilon:
                action = env.action_space.sample()
            else:
                action = np.argmax(q_table[state_d])

            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            next_state_d = discretize_state(next_state, bins)

            # Обновление Q-таблицы
            if done:
                q_table[state_d + (action,)] += alpha * (reward - q_table[state_d + (action,)])
            else:
                best_next_action = np.argmax(q_table[next_state_d])
                td_target = reward + gamma * q_table[next_state_d + (best_next_action,)]
                td_error = td_target - q_table[state_d + (action,)]
                q_table[state_d + (action,)] += alpha * td_error

            state_d = next_state_d

        if epsilon > epsilon_min:
            epsilon *= epsilon_decay

        if ep % 500 == 0:
            print(f"Episode {ep}, Epsilon: {epsilon:.3f}")

    env.close()
    return q_table
```

```python
def test_agent(q_table, episodes=5):
    env = gym.make('CartPole-v1', render_mode="human")  # Для визуализации
    for ep in range(episodes):
        state, _ = env.reset()
        state_d = discretize_state(state, bins)
        done = False
        total_reward = 0

        while not done:
            action = np.argmax(q_table[state_d])
            state, reward, terminated, truncated, _ = env.step(action)
            state_d = discretize_state(state, bins)
            done = terminated or truncated
            total_reward += reward

        print(f"Episode {ep + 1}, Total Reward: {total_reward}")

    env.close()
```

```python
q_table = train_cartpole()
test_agent(q_table)
```
