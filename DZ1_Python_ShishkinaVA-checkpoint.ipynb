```python
import math
```

```python
def solution_equation(coefficients, deg):
    if deg == 0: 
        if abs(coefficients[0]) < 1e-12:
            return ["Бесконечное множество решений"]
        else:
            return ["Уравнение не имеет решений"]
    elif deg == 1:
        return solution_linear(coefficients[0], coefficients[1])
    elif deg == 2:
        return solution_quadratic(coefficients[0], coefficients[1], coefficients[2])
    elif deg == 3:
        return solution_cubic(coefficients[0], coefficients[1], coefficients[2], coefficients[3])
    elif deg == 4:
        return solution_quartic(coefficients[0], coefficients[1], coefficients[2], coefficients[3], coefficients[4])
    else: 
        return ["Степень уравнения должна быть < 5"]
```

Решение линейного уравнения.

```python
def solution_linear(a, b):
    if a == 0:
        if b == 0:
            return ["Бесконечное множество решений"]
        else:
            return ["Уравнение не имеет решения"]
    return [-b/a]
```

Решение уравнения второй степени.

```python
def solution_quadratic(a, b, c):
    if a == 0:
        return solution_linear(b, c)
    discriminant = b*b - 4*a*c
    if discriminant > 0:
        sqrtDiscriminant = discriminant**0.5
        x1 = (-b + sqrtDiscriminant) / (2*a)
        x2 = (-b - sqrtDiscriminant) / (2*a)
        return [x1, x2]
    elif discriminant == 0:
        return [-b / (2*a)]
    else:
        sqrtDiscriminant = discriminant**0.5
        x1 = (-b + sqrtDiscriminant) / (2*a)
        x2 = (-b - sqrtDiscriminant) / (2*a)
        return [x1, x2]
```

Решение уравнения третьей степени.

```python
def solution_cubic(a, b, c, d):
    if a == 0:
        return solution_quadratic(b, c, d)

    p = b / a
    q = c / a
    r = d / a

    m = q - p**2 / 3
    n = 2*p**3/27 - p*q/3 + r
    discriminant = (n/2)**2 + (m/3)**3

    if discriminant < 0:
        u = (-n/2)**(1/3) if n >= 0 else -((-n/2)**(1/3))
        y1 = 2*u
        y2 = -u
        y3 = -u
    elif discriminant > 0:
        sqrtD = discriminant**0.5
        u = -n/2 + sqrtD
        v = -n/2 - sqrtD
        u = u**(1/3) if u>=0 else -((-u)**(1/3))
        v = v**(1/3) if v>=0 else -((-v)**(1/3))
        y1 = u + v

        y2 = -(u + v)/2 + (u - v)*3**0.5/2j
        y3 = -(u + v)/2 - (u - v)*3**0.5/2j
    else:
        rho = ((-m/3)**3)**0.5
        theta = math.acos(-n / (2 * rho))
        y1 = 2 * rho**(1/3)*math.cos(theta/3)
        y2 = 2 * rho**(1/3)*math.cos((theta + 2*math.pi)/3)
        y3 = 2 * rho**(1/3)*math.cos((theta + 4*math.pi)/3)

    x1 = y1 - p/3
    x2 = y2 - p/3
    x3 = y3 - p/3

    return [x1, x2, x3]
```

Решение уравнения четвертой степени.

```python
def solution_quartic(a, b, c, d, e):
    if a == 0:
        return solution_cubic(b, c, d, e)
    
    p = b / a
    q = c / a
    r = d / a
    s = e / a

    A = 1
    B = -q
    C = p*r - 4*s
    D = -p**2*s + 4*q*s - r**2

    roots_y = solution_cubic(A, B, C, D)
    y = None
    for root in roots_y:
        if isinstance(root, (int, float)) or (isinstance(root, complex) and root.imag == 0):
            y = root.real if isinstance(root, complex) else root
            break
    if y is None:
        y = roots_y[0].real

    m = (p**2/4 - q + y)**0.5 if (p**2/4 - q + y) >= 0 else (p**2/4 - q + y)**0.5
    n = (p*y/2 - r) / m if m != 0 else 0
    
    roots1 = solution_quadratic(1, p/2 + m, y/2 + n)
    roots2 = solution_quadratic(1, p/2 - m, y/2 - n)
    
    return roots1 + roots2
```

Функция печати уравнения.

```python
def print_equation(coefficients):
    terms = []
    n = len(coefficients) - 1
    for i, coef in enumerate(coefficients):
        power = n - i
        if abs(coef) < 1e-12:
            continue
            
        if power == 0:
            term = f"{coef:.6g}"
        elif power == 1:
            if abs(coef - 1) < 1e-12:
                term = "x"
            elif abs(coef + 1) < 1e-12:
                term = "-x"
            else:
                term = f"{coef:.6g}*x"
        else:
            if abs(coef - 1) < 1e-12:
                term = f"x^{power}"
            elif abs(coef + 1) < 1e-12:
                term = f"-x^{power}"
            else:
                term = f"{coef:.6g}*x^{power}"
        terms.append(term)
    
    if not terms:
        print("0 = 0")
    else:
        equation = " + ".join(terms).replace("+ -", "- ")
        print(f"{equation} = 0")
```
