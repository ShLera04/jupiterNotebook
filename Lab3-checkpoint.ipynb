```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Определяем функцию
def f(xy):
    x, y = xy
    return (x - 1)**2 + (y - 2)**2

# Начальная точка для поиска минимума
initial_guess = [0, 0]

# Выполняем оптимизацию
result = minimize(f, initial_guess)

# Получаем координаты минимума
min_x, min_y = result.x
min_value = result.fun

print(f"Локальный минимум находится в точке: ({min_x:.2f}, {min_y:.2f})")
print(f"Значение функции в этой точке: {min_value:.2f}")

# Создаем сетку для отрисовки контуров
x = np.linspace(-1, 3, 400)
y = np.linspace(0, 4, 400)
X, Y = np.meshgrid(x, y)
Z = f([X, Y])

# Отрисовываем контуры функции
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20)
plt.clabel(contour, inline=True, fontsize=8)
plt.plot(min_x, min_y, 'ro')  # Отмечаем локальный минимум
plt.title('Контуры функции и локальный минимум')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Определяем функцию
def f(xy):
    x, y = xy
    return x**2 + y**2

# Начальная точка для поиска минимума
initial_guess = [1, 1]

# Выполняем оптимизацию
result = minimize(f, initial_guess)

# Получаем координаты минимума
min_x, min_y = result.x
min_value = result.fun

print(f"Локальный минимум находится в точке: ({min_x:.2f}, {min_y:.2f})")
print(f"Значение функции в этой точке: {min_value:.2f}")

# Создаем сетку для отрисовки контуров
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = f([X, Y])

# Отрисовываем контуры функции
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20)
plt.clabel(contour, inline=True, fontsize=8)
plt.plot(min_x, min_y, 'ro')  # Отмечаем локальный минимум
plt.title('Контуры функции f(x, y) = x^2 + y^2 и локальный минимум')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize

# Определяем функцию
def f(xy):
    x, y = xy
    return np.cos(np.sqrt(x**2 + y**2)) * np.sin(x)

# Начальная точка для поиска минимума
initial_guess = [0, 0]

# Находим локальный минимум
result = minimize(f, initial_guess)

# Получаем координаты минимума
min_x, min_y = result.x
min_z = f(result.x)

# Создаем сетку значений
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f([X, Y])

# Создаем 3D график
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Рисуем поверхность
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.7)

# Отмечаем локальный минимум
ax.scatter(min_x, min_y, min_z, color='red', s=100, label='Локальный минимум')
ax.text(min_x, min_y, min_z, f'Min ({min_x:.2f}, {min_y:.2f}, {min_z:.2f})', color='red')

# Настройки графика
ax.set_title('График функции с локальным минимумом')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.legend()

# Создаем линии уровня
plt.figure(figsize=(10, 8))
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contour)  # Добавляем цветовую шкалу

# Отмечаем локальный минимум на графике линий уровня
plt.scatter(min_x, min_y, color='red', s=100, label='Локальный минимум')
plt.title('Линии уровня функции с локальным минимумом')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```
