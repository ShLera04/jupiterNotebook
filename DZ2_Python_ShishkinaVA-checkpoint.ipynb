```python
def niceEnd(n):
    if n in [11,12,13,14]:
        return f"{n} лет"  
    if n%10 in [2,3,4]:
        return f"{n} года"
    if n%10==1:
        return f"{n} год"
    return f"{n} лет"
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def getInfo(self):
        return f"Меня зовут {self.name}. Мне {niceEnd(self.age)}."
    def __str__(self):
        return self.getInfo()
```

```python
class Patient(Person):
    def __init__(self, name, age, symptoms, inСonsciousness=True, inСritical=False):
        super().__init__(name, age)
        self.symptoms = symptoms
        self.inСonsciousness = inСonsciousness
        self.inСritical = inСritical
        self.diagnosis = None
        self.medicine = None
    def getInfo(self):
        return f"Пациент: {self.name}, возраст: {niceEnd(self.age)}, симптомы: {', '.join(self.symptoms)}, диагноз: {self.diagnosis}, препарат: {self.medicine}."
    def __str__(self):
        return self.getInfo()
    def isReadyForCheck(self):
        return self.inСonsciousness and not self.inСritical
```

```python
class Doctor(Person):
    def __init__(self, name, age, specialty):
        super().__init__(name, age)
        self.specialty = specialty
    def examinePatient(self, patient):
        print(f"Врач: {self.name} ({self.specialty}) осматривает пациента.")
```

```python
class SymptomChecker:
    def __init__(self):
        self.rules = {
            frozenset(["температура", "кашель", "слабость", "головная боль", "насморк"]): "грипп",
            frozenset(["зуд", "высыпания", "отек", "слезотечение"]): "аллергия",
            frozenset(["потеря сознания", "тошнота", "рвота", "звон в ушах", "амнезия"]): "сотрясение мозга",
            frozenset(["головная боль", "головокружение", "нарушение координации, речи", "асимметрия лица"]): "инсульт",
            frozenset(["температура", "кашель", "одышка", "боль в грудной клетке"]): "пневмония"}
    def diagnose(self, doctor, patient):
        try:
            if not patient.isReadyForCheck():
                raise RuntimeError("Пациенту нужна экстренная помощь!")
                
            patientSymptomps = set(patient.symptoms)
            match patientSymptomps:
                case s if frozenset(s) in self.rules:
                    diagnosis = self.rules[frozenset(s)]
                    print(f"Точный диагноз для пациента: {diagnosis}.")
                    patient.diagnosis = diagnosis
                case s if {"температура", "кашель", "слабость"}.issubset(s):
                    print("Подозрение на грипп.")
                    patient.diagnosis = "Грипп (требуются анализы)"
                case s if {"зуд", "высыпания"}.issubset(s):
                    print("Подозрение на аллергию.")
                    patient.diagnosis = "Аллергия (требуются анализы)"
                case s if {"головная боль", "нарушение координации, речи"}.issubset(s):
                    print("Подозрение на инсульт.")
                    patient.diagnosis = "Инсульт (требуются доп. обследования)"
                case s if {"одышка", "боль в грудной клетке"}.issubset(s):
                    print("Подозрение на пневмонию.")
                    patient.diagnosis = "Пневмония (требуются доп. обследования)"
                case s if {"звон в ушах", "амнезия"}.issubset(s):
                    print("Подозрение на сотрясение мозга.")
                    patient.diagnosis = "Сотрясение мозга (требуются доп. обследования)."
                case _:
                    print("Симптомы не ясны, требуется дополнительное обследование.")
                    patient.diagnosis = "непоставлен"
        except RuntimeError as e:
            print("Пациент направлен в реанимацию.")
            patient.diagnosis = "непоставлен"
```

```python
class MedicalDrug:
    def __init__(self):
        self.medicines = {
            "грипп": "терафлю",
            "аллергия": "супрастин",
            "сотрясение мозга": "кавинтон",
            "инсульт": "пентоксифиллин",
            "пневмония": "азитромицин"}
    def appointment(self, patient):
        try:
            if patient.diagnosis not in self.medicines:
                raise KeyError(f"Лекарство не найдено для диагноза: {patient.diagnosis}.")
            patient.medicine = self.medicines[patient.diagnosis]
            print(f"Назначен препарат: {patient.medicine}.")
        except KeyError as e:
            print("Требуется наблюдение, препарат не назначен.")
            patient.medicine = "неназначен"
```

```python
doctor = Doctor("Петрова Анна Ивановна", 35, "терапевт")
checker = SymptomChecker()
drug = MedicalDrug()

patient = Patient("Макаров Иван Иванович", 25, ["температура", "кашель", "слабость"])
print(f"\n---Прием пациента: {patient.name} ----")

doctor.examinePatient(patient)
checker.diagnose(doctor, patient)
drug.appointment(patient)

print("\n")
print(patient)
```

```python
import random
```

```python
symptoms = ["температура", "кашель", "слабость", "головная боль", "насморк","зуд", "высыпания", "отек", "слезотечение", 
            "потеря сознания", "тошнота", "рвота", "звон в ушах", "амнезия", "головокружение", "нарушение координации, речи",
            "асимметрия лица","одышка", "боль в грудной клетке"]

def generatePatient(count):
    for i in range(count):
        name = f"Пациент_{i+1}"
        age = random.randint(5, 100)
        countSymptoms = random.randint(1, 5)
        symptomsPerson = random.sample(symptoms, countSymptoms)
        yield Patient(name, age, symptomsPerson)

for patient in generatePatient(3):
    print(f"Пациент: {patient.name}, возраст: {niceEnd(patient.age)}, симптомы: {', '.join(patient.symptoms)}")
```
