```python
def niceEnd(n):
    if n%100 in [11,12,13,14]:
        return f"{n} лет"  
    if n%10 in [2,3,4]:
        return f"{n} года"
    if n%10==1:
        return f"{n} год"
    return f"{n} лет"
```

Функция niceEnd(n) формирует возраст человека с правильным склонением.

```python
class Person:
    def __init__(self, name, age, sex="не определен"):
        self.name = name
        self.age = age
        self.sex = sex
    def getInfo(self):
        return f"Меня зовут {self.name}. Мне {niceEnd(self.age)}. Пол: {self.sex}"
    def __str__(self):
        return self.getInfo()
```

Класс Person представляет человека с его именем, возрастом и полом.

Методы:
1. init(...) - инициализация полей
2. getInfo(...) - получение информации об объекте
3. str(...) - делегирует вывод функции getInfo(...)

```python
class Patient(Person):
    def __init__(self, name, age, symptoms, sex="не определен", history=None, outcome=None):
        super().__init__(name, age, sex)
        self.symptoms = symptoms
        self.history = history if history is not None else []
        self.diagnosis = None
        self.treatment = None
        self.outcome = outcome
    def getInfo(self):
        hist_str = ", ".join(self.history) if self.history else "нет"
        return (f"Пациент: {self.name}, возраст: {niceEnd(self.age)}, пол: {self.sex}, симптомы: {', '.join(self.symptoms)}, "
        f"история болезни: {hist_str}, диагноз: {self.diagnosis}, лечение: {self.treatment}, результат: {self.outcome}.")
    def __str__(self):
        return self.getInfo()
    def __repr__(self):
        return (f"Patient(name='{self.name}', age={self.age}, sex: {self.sex}, symptoms={self.symptoms}, "
        f"history: {self.history}, diagnosis={self.diagnosis}, treatment: {self.treatment}, outcome: {self.outcome}")
    def __eq__(self, other):
        if not isinstance(other, Patient):
            return NotImplemented
        return self.name == other.name and self.age == other.age and self.sex == other.sex
    def __lt__(self, other):
        if not isinstance(other, Patient):
            return NotImplemented
        return self.age < other.age
    def __le__(self, other):
        if not isinstance(other, Patient):
            return NotImplemented
        return self.age <= other.age
    def __gt__(self, other):
        if not isinstance(other, Patient):
            return NotImplemented
        return self.age > other.age
    def __ge__(self, other):
        if not isinstance(other, Patient):
            return NotImplemented
        return self.age >= other.age
    def getName(self):
        return self.name
    def getAge(self):
        return self.age
```

Класс Patient наследуется от класса Person. Он моделирует пациента с симптомами, историей болезни, диагнозом, лечением и результатом лечения.

Методы:
1. init(...) - инициализация полей
2. getInfo(...) - получение информации об объекте
3. str(...), repr(...) - делегируют вывод функции getInfo(...)
4. eq(...), lt(...), le(...), gt(...), ge(...) - операторы сравнения
5. getName(...), getAge(...)  - функции, для доступа к полям класса

```python
class Diagnosis:
    def __init__(self, name, severity="умеренный"): #'легкий', 'умеренный', 'тяжелый'
        self.name = name
        self.severity = severity
    def __str__(self):
        return f"Diagnosis(name = '{self.name}', severity = '{self.severity}')"
```

Класс Diagnosis представляет медицинский диагноз с названием и степенью тяжести.

Степень тяжести: легкий, умеренный, тяжелый

Методы:
1. init(...) - инициализация полей
2. str(...) - возвращает описание объекта

```python
class Treatment:
    def __init__(self, name, category="назначение препаратов"): # 'назначение препаратов', 'медицинская процедура', 'операция'
        self.name = name
        self.category = category
    def __str__(self):
        return f"Treatment(name = '{self.name}', category = '{self.category}')"
```

Класс Treatment представляет медицинское лечение с названием и категорией.

Категория: назначение препаратов, медицинская процедура, операция

Методы:
1. init(...) - инициализация полей
2. str(...) - возвращает описание объекта

```python
class TreatmentOutcome:
    def __init__(self, result, duration, sideEffects=None):
        # result: 'выздоровел', 'улучшение состояния', 'без изменений', 'ухудшение состояния', 'умер'
        self.result = result
        self.duration = duration
        self.sideEffects = sideEffects if sideEffects is not None else []
    def __repr__(self):
        return f"TreatmentOutcome(result='{self.result}', duration={self.duration}, sideEffects={self.sideEffects})"
```

Класс TreatmentOutcome хранит исход лечения: результат, длительность лечения и побочные эффекты.

Методы:
1. init(...) - инициализация полей
2. repr(...) - возвращает описание объекта

```python
class DiagnosisPredictor:
    def __init__(self):
        self.rules = {
            frozenset(["температура", "кашель", "слабость", "головная боль", "насморк"]): (Diagnosis("грипп", "умеренный"), Treatment("терафлю", "назначение препаратов")),
            frozenset(["зуд", "высыпания", "отек", "слезотечение"]): (Diagnosis("аллергия", "легкий"), Treatment("супрастин", "назначение препаратов")),
            frozenset(["потеря сознания", "тошнота", "рвота", "звон в ушах", "амнезия"]): (Diagnosis("сотрясение мозга", "тяжелый"), Treatment("кавинтон", "назначение препаратов")),
            frozenset(["головная боль", "головокружение", "нарушение координации, речи", "асимметрия лица"]): (Diagnosis("инсульт", "тяжелый"), Treatment("пентоксифилин", "назначение препаратов")),
            frozenset(["температура", "кашель", "одышка", "боль в грудной клетке"]):  (Diagnosis("пневмония", "тяжелый"), Treatment("азитромицин", "medication"))}
    def predictAndPrescribe(self, patient):
        diagnosis = None
        treatment = None
        patientSymptoms = set(patient.symptoms)
        age = patient.age
    
        match (patientSymptoms, age):
            case (s, _) if frozenset(s) in self.rules:
                d, t = self.rules[frozenset(s)]
                diagnosis, treatment = d, t
                #print(f"Точный диагноз и лечение для {patient.name}: {diagnosis}, {treatment}.")
                patient.diagnosis = diagnosis
                patient.treatment = treatment
            case (s, a) if {"температура", "кашель", "слабость"}.issubset(s) and a > 65:
                diagnosis = Diagnosis("грипп (у пожилого)", "тяжелый")
                treatment = Treatment("терафлю и наблюдение", "назначение препаратов")
                #print(f"Подозрение на грипп у пожилого пациента {patient.name}.")
                patient.diagnosis = diagnosis
                patient.treatment = treatment
            case (s, _) if {"температура", "кашель", "слабость"}.issubset(s):
                diagnosis = Diagnosis("грипп (требуются анализы)", "умеренный")
                treatment = Treatment("терафлю", "назначение препаратов")
                #print("Подозрение на грипп.")
                patient.diagnosis = diagnosis
                patient.treatment = treatment
            case (s, _) if {"зуд", "высыпания"}.issubset(s):
                diagnosis = Diagnosis("аллергия (требуются анализы)", "легкий")
                treatment = Treatment("супрастин", "назначение препаратов")
                #print("Подозрение на аллергию.")
                patient.diagnosis = diagnosis
                patient.treatment = treatment
            case (s, _) if ({"головная боль", "нарушение координации, речи"}.issubset(s) or
                            {"асимметрия лица", "головокружение"}.issubset(s)):
                diagnosis = Diagnosis("подозрение на инсульт", "тяжелый")
                treatment = Treatment("пентоксифиллин и срочная диагностика", "назначение препаратов")
                #print("Подозрение на инсульт.")
                patient.diagnosis = diagnosis
                patient.treatment = treatment
            case (s, _) if {"одышка", "боль в грудной клетке"}.issubset(s):
                diagnosis = Diagnosis("подозрение на пневмонию", "тяжелый")
                treatment = Treatment("азитромицин и рентген", "назначение препаратов")
                #print("Подозрение на пневмонию.")
                patient.diagnosis = diagnosis
                patient.treatment = treatment
            case (s, _) if {"звон в ушах", "амнезия"}.issubset(s):
                diagnosis = Diagnosis("подозрение на сотрясение мозга", "тяжелый")
                treatment = Treatment("кавинтон и МРТ", "назначение препаратов")
                #print("Подозрение на сотрясение мозга.")
                patient.diagnosis = diagnosis
                patient.treatment = treatment
            case _:
                diagnosis = Diagnosis("непоставлен", "неназначен")
                treatment = Treatment("неназначен", "неназначен")
                #print("Симптомы не ясны, требуется дополнительное обследование.")
                patient.diagnosis = diagnosis
                patient.treatment = treatment
        return diagnosis, treatment
```

Класс DiagnosisPredictor представляет собой систему для постановки диагноза и назначения лечения на основе симптомов пациента. она анализирует симптомы пациента и сопоставляет их с заданными правилами.

Методы:
1. init(...) - создает словарь с точными наборами симптомов и диагнозом, лечением.
2. predictAndPrescribe(...) - функция анализирует совпадение симптомов со словарем правил, устанавливает диагноз и лечение.

```python
import random

names = [f"Пациент_{i}" for i in range(1, 1001)]
symptomsPool = ["температура", "кашель", "слабость", "головная боль", "насморк",
                 "зуд", "высыпания", "отек", "слезотечение", "потеря сознания",
                 "тошнота", "рвота", "звон в ушах", "амнезия", "головокружение",
                 "нарушение координации, речи", "асимметрия лица", "одышка", "боль в грудной клетке"]
sexes = ["мужской", "женский"]
resultsPool = ["выздоровел", "улучшение состояния", "без изменений", "ухудшение состояния", "умер"]

def generatePatientRecords(count, predictor):
    for i in range(count):
        try:
            name = random.choice(names)
            age = random.randint(5, 100)
            sex = random.choice(sexes)
            countSymptoms = random.randint(1, 5)
            symptoms = random.sample(symptomsPool, countSymptoms)

            patient = Patient(name, age, symptoms, sex=sex)

            predictor.predictAndPrescribe(patient)

            baseDuration = random.randint(3, 30)
            sideEffects = random.sample(["головная боль", "тошнота", "сонливость", "бессонница"], random.randint(0, 2))
            resultProb = random.random()
            if resultProb < 0.55:
                result = "выздоровел"
            elif resultProb < 0.75:
                result = "улучшение состояния"
            elif resultProb < 0.85:
                result = "без изменений"
            elif resultProb < 0.95:
                result = "ухудшение состояния"
            else:
                result = "умер"

            outcome = TreatmentOutcome(result, baseDuration, sideEffects)
            patient.outcome = outcome

            yield patient

        except Exception as e:
            print(f"Ошибка генерации записи {i}: {e}")
            continue
```

Функция generatePatientRecords(...) генерирует синтетические записи пациентов, заполняя их случайными данными. Она создает заданное количество пациентов с определенными симптомами, автоматически ставит диагноз и назначает лечение с использованием метода predictAndPrescribe(...), генерирует исход лечения.

Задаются данные для генерации, создается пациент, он передается в predictor - систему для постановки диагноза, затем формируется исход на основании полученных данных.

```python
from functools import reduce
from collections import Counter
import operator
import traceback

def analyzeTreatmentEffectiveness(records):
    print("\n--- Анализ эффективности лечения ---")
    try:
        treatmentResults = {}
        for patient in records:
            treatmentName = str(patient.treatment)
            outcomeResult = patient.outcome.result
            if treatmentName not in treatmentResults:
                treatmentResults[treatmentName] = []
            treatmentResults[treatmentName].append(outcomeResult)

        effectiveness = {}
        for treatment, outcomes in treatmentResults.items():
            successful = sum(1 for o in outcomes if o in ['выздоровел', 'улучшение состояния'])
            total = len(outcomes)
            if total > 0:
                effectiveness[treatment] = round((successful / total) * 100, 2)
            else:
                effectiveness[treatment] = 0

        print("Эффективность лечения (%):")
        for treatment, eff in sorted(effectiveness.items(), key=operator.itemgetter(1), reverse=True):
             print(f"  {treatment}: {eff}%")

        return effectiveness

    except ZeroDivisionError:
        print("Ошибка: Деление на ноль при подсчете эффективности.")
        return {}
    except Exception as e:
        print(f"Ошибка при анализе эффективности лечения: {e}")
        traceback.print_exc()
        return {}


def identifyRiskFactors(records):
    print("\n--- Анализ факторов риска (частые симптомы при плохом исходе) ---")
    try:
        badOutcomePatients = list(filter(lambda p: p.outcome.result in ['ухудшение состояния', 'умер'], records))

        allRiskSymptoms = []
        symptomLists = map(lambda p: p.symptoms, badOutcomePatients)
        for symptomList in symptomLists:
            allRiskSymptoms.extend(symptomList)

        riskCounter = Counter(allRiskSymptoms)
        mostCommonRisks = riskCounter.most_common(5)

        print("Топ-5 симптомов, связанных с плохим исходом:")
        for symptom, count in mostCommonRisks:
            print(f"{symptom}: {count} случаев")

        return riskCounter

    except Exception as e:
        print(f"Ошибка при анализе факторов риска: {e}")
        traceback.print_exc()
        return Counter()


def analyzePatientDemographics(records):
    print("\n--- Анализ по возрасту ---")
    try:
        ageGroups = {"<30": [], "30-50": [], "50-70": [], ">70": []}
        for p in records:
            age = p.age
            if age < 30:
                group = "<30"
            elif 30 <= age < 50:
                group = "30-50"
            elif 50 <= age < 70:
                group = "50-70"
            else:
                group = ">70"
            ageGroups[group].append(p)

        ageOutcomes = {}
        for group, patients in ageGroups.items():
            outcomesInGroup = list(map(lambda p: p.outcome.result, patients))
            outcomeCounts = dict(Counter(outcomesInGroup))
            ageOutcomes[group] = outcomeCounts
            print(f"Группа {group}: {outcomeCounts}")

        return ageOutcomes

    except Exception as e:
        print(f"Ошибка при анализе по возрасту: {e}")
        traceback.print_exc()
        return {}
```

Функция analyzeTreatmentEffectiveness(...) группирует исходы по названиям лечения, вычисляет процент успешных исходов, формирует рейтинг эффективности лечения.

Функция identifyRiskFactors(...) фильтрует пациентов с плохими исходами, считает частоту симптомов и выводит 5 самых рискованных.

Функция analyzePatientDemographics(...) делит пациентов на четыре возрастные группы, подсчитывает исходы в каждой группе, выводит статистику по ним.

```python
import random
import threading

random.seed(42)

print("--- Начало работы системы анализа ---")

predictor = DiagnosisPredictor()

numRecords = 500
print(f"Генерация {numRecords} записей пациентов.")

records = list(generatePatientRecords(numRecords, predictor))

print(f"Сгенерировано {len(records)} записей.")
if records:
    print("Пример сгенерированной записи:")
    print(repr(records[0]))
    print()

print("--- Запуск анализа в потоках ---")

threads = []

t1 = threading.Thread(target=lambda: analyzeTreatmentEffectiveness(records))
t2 = threading.Thread(target=lambda: identifyRiskFactors(records))
t3 = threading.Thread(target=lambda: analyzePatientDemographics(records))

threads.extend([t1, t2, t3])

for t in threads:
    t.start()

for t in threads:
    t.join()

patientsWithFlu = list(filter(lambda p: p.diagnosis and p.diagnosis.name == "грипп", records))
print(f"Количество пациентов с диагнозом 'грипп': {len(patientsWithFlu)}")
```
