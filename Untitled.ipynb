```python
def degree1(coeff_list):
    answer = []

    if coeff_list[0] == 0 and coeff_list[1] == 0:
        answer.append(float('inf'))
    else:
        answer.append(-coeff_list[1]/coeff_list[0])

    return answer
```

```python
def degree2(coeff_list):
    answer = []

    D = coeff_list[1]**2 - 4*coeff_list[0]*coeff_list[2]
    if D > 0:
        x1 = (-coeff_list[1] - D**0.5) / (2*coeff_list[0])
        x2 = (-coeff_list[1] + D**0.5) / (2*coeff_list[0])
        answer.append(x1)
        answer.append(x2)
    elif D == 0:
        x = -coeff_list[1] / 2*coeff_list[0]
        answer.append(x)

    return answer
```

```python
def polyval(coeff_list, x):
    result = 0
    for coeff in coeff_list:
        result = result * x + coeff
    return result

def newton_root(coeff_list, x0=0.0, max_iter=100, tol=1e-7):
    result = None
    start_values = [x0, 1.0, -1.0, 2.0, -2.0, 10.0, -10.0]

    for start in start_values:
        x = start

        for _ in range(max_iter):
            fx = polyval(coeff_list, x)

            if abs(fx) < tol:
                result = x
                break

            deriv_coeffs = []
            n = len(coeff_list)
            for i in range(n - 1):
                deriv_coeffs.append(coeff_list[i] * (n - i - 1))

            fpx = polyval(deriv_coeffs, x)
            if abs(fpx) < tol:
                break

            x_new = x - fx / fpx
            if abs(x_new - x) < tol:
                result = x_new
                break
            x = x_new

        if result is not None:
            break

    return result

def synthetic_division(coeff_list, root):
    quotient = []
    current = 0
    for coeff in coeff_list:
        current = current * root + coeff
        quotient.append(current)
    return quotient[:-1]
```

```python
def degree3(coeff_list):
    if coeff_list[3] == 0:
        return degree2(coeff_list[1::])

    root = newton_root(coeff_list)
    if root is None:
        return []

    quotient = synthetic_division(coeff_list, root)
    roots = degree2(quotient)
    roots.append(root)
    return roots
```

```python
def degree4(coeff_list):
    if coeff_list[4] == 0:
        return degree3(coeff_list[1::])

    root = newton_root(coeff_list)
    if root is None:
        return []

    quotient = synthetic_division(coeff_list, root)
    roots = degree3(quotient)
    roots.append(root)
    return roots
```

```python
def equation_view(coeff_list, degree):
    coeff_list = coeff_list[::-1]
    equation = ''
    for index in range(len(coeff_list) - 1, -1, -1):
        elem = coeff_list[index]
        #print(index, " : ", elem)

        if index != degree and elem > 0 and equation != '':
            equation += '+'

        if index == 1:
            if elem != 0 and elem != 1 and elem != -1:
                equation += str(elem) + 'x'
            elif elem == 1:
                equation += 'x'
            elif elem == -1:
              equation += '-x'

        elif index == 0 and elem != 0:
                equation += str(elem)

        else:
            if elem != 0 and elem != 1 and elem != -1:
                equation += str(elem) + 'x^' + str(index)
            elif elem == 1:
                equation += 'x^' + str(index)
            elif elem == -1:
              equation += '-x^' + str(index)

    if equation == '':
        equation = '0=0'
    else:
        equation += '=0'
    print(f"Вид уравнения: {equation}")
```
