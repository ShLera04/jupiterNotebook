Метод последовательного квадратичного программирования (Sequential Quadratic Programming, SQP) — один из наиболее распространённых и эффективных оптимизационных алгоритмов общего назначения, основной идеей которого является последовательное решение задач квадратичного программирования, аппроксимирующих данную задачу оптимизации. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from mpl_toolkits.mplot3d import Axes3D
```

Задача №1.

Рассматриваемая функция: Q(x,y) = 2(x)^2 + 5(y)^2

Ограничения: 

g1(x,y) = y + 0.5 * x + 0.5

g2(x, y) = y + 0.5 * x


```python
def Q(x):
    return 2 * x[0]**2 + 5 * x[1]**2

def g1(x):
    return -(x[1] + 0.5 * x[0] + 0.5)

def g2(x):
    return -(x[1] - 0.5 * x[0])

x0 = [-1.5, -1.5]

constraints = (
    {'type': 'ineq', 'fun': g1},
    {'type': 'ineq', 'fun': g2}
)

result = minimize(Q, x0, method='SLSQP', constraints=constraints)

print("Решение:", result.x)
print("Значение функции в точке решения:", result.fun)

x = np.linspace(-1.8, 1.8, 400)
y = np.linspace(-2, 1, 400)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

g1_value = -(Y + 0.5 * X + 0.5)
g2_value = -(Y - 0.5 * X)

fig, ax = plt.subplots(figsize=(8, 6))

contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour, label='Q(x)')

ax.contour(X, Y, g1_value, [0], colors='red', linewidths=3)
ax.contour(X, Y, g2_value, [0], colors='green', linewidths=3)

feasible_region = (g1_value >= 0) & (g2_value >= 0)
ax.contourf(X, Y, feasible_region, levels=[0.5, 1], colors='lightblue', alpha=0.5)

ax.scatter(result.x[0], result.x[1], color='blue', label='Оптимальная точка')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('Рельеф функции и допустимая область')
ax.legend()
plt.show()
```

Визуализируем в 3D.

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

ax.contour(X, Y, g1_value, [0], zdir='z', offset=0, colors='red', linewidths=3)
ax.contour(X, Y, g2_value, [0], zdir='z', offset=0, colors='green', linewidths=3)

feasible_region = (g1_value >= 0) & (g2_value >= 0)
ax.contourf(X, Y, feasible_region, zdir='z', offset=0, levels=[0.5, 1], colors='lightblue', alpha=0.5)

ax.scatter(result.x[0], result.x[1], result.fun, color='blue', label='Оптимальная точка', zorder=5)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Q(x)')
ax.set_title('3D визуализация функции и допустимой области')
ax.legend()
plt.show()
```

Задача №2.

Рассматриваемая функция: Q(x,y) = cos(0.5x)*sin(0.5y)

Ограничения: 

g1(x) = x - 2

g2(x, y) = -x + y - 2

g3(x,y) = -x - y + 2 


```python
def Q1(x):
    return np.cos(0.5 * x[0]) * np.sin(0.5 * x[1])

def g11(x):
    return -(x[0] - 2)

def g22(x):
    return -(-x[0] + x[1] - 2)

def g33(x):
    return -(-x[0] - x[1] + 2)

x02 = [-1.5, -1.5]

constraints = (
    {'type': 'ineq', 'fun': g11},
    {'type': 'ineq', 'fun': g22},
    {'type': 'ineq', 'fun': g33}
)

result = minimize(Q1, x02, method='SLSQP', constraints=constraints)

print("Решение:", result.x)
print("Значение функции в точке решения:", result.fun)

x = np.linspace(-1, 10, 100)
y = np.linspace(-5, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(0.5 * X) * np.sin(0.5 * Y)

g11_value = -(X - 2)
g22_value = -(-X + Y - 2)
g33_value = -(-X - Y + 2)

fig, ax = plt.subplots(figsize=(10, 8))

contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour, label='Q(x)')

ax.contour(X, Y, g11_value, [0], colors='red', linewidths=3)
ax.contour(X, Y, g22_value, [0], colors='green', linewidths=3)
ax.contour(X, Y, g33_value, [0], colors='blue', linewidths=3)

feasible_region = (g11_value >= 0) & (g22_value >= 0) & (g33_value >= 0)
ax.contourf(X, Y, feasible_region, levels=[0.5, 1], colors='lightblue', alpha=0.5)

ax.scatter(result.x[0], result.x[1], color='black', label='Оптимальная точка')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('Рельеф функции и допустимая область')
ax.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.5)

ax.contour(X, Y, g11_value, [0], zdir='z', offset=0, colors='red', linewidths=3)
ax.contour(X, Y, g22_value, [0], zdir='z', offset=0, colors='green', linewidths=3)
ax.contour(X, Y, g33_value, [0], zdir='z', offset=0, colors='blue', linewidths=3)


feasible_region = (g11_value >= 0) & (g22_value >= 0) & (g33_value >= 0)
ax.contourf(X, Y, feasible_region, zdir='z', offset=0, levels=[0.5, 1], colors='lightblue', alpha=0.5)

ax.scatter(result.x[0], result.x[1], result.fun, color='black', label='Оптимальная точка')
ax.set_zlim(-1, 1)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Q(x)')
ax.set_title('3D визуализация функции и допустимой области')
ax.legend()
plt.show()
```

Задача №3.

Рассматриваемая функция: Q(x,y) = sin(xy)

Ограничения: 

g1(x) = -x + 1

g2(x) = -5 + x

g3(y) = -y + 1

g4(y) = -5 + y

```python
def Q3(x):
    return np.sin(x[0]*x[1])

def g111(x):
    return (x[0] - 1)

def g222(x):
    return (5 - x[0])

def g333(x):
    return (x[1] - 1)

def g444(x):
    return (5 - x[1])

x03 = [0, 0]

constraints = (
    {'type': 'ineq', 'fun': g111},
    {'type': 'ineq', 'fun': g222},
    {'type': 'ineq', 'fun': g333},
    {'type': 'ineq', 'fun': g444}
)

result = minimize(Q3, x03, method='SLSQP', constraints=constraints)

print("Решение:", result.x)
print("Значение функции в точке решения:", result.fun)

x = np.linspace(-5, 6, 100)
y = np.linspace(-5, 6, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X*Y)

g111_value = (X - 1)
g222_value = (5 - X)
g333_value = (Y - 1)
g444_value = (5 - Y)

fig, ax = plt.subplots(figsize=(10, 8))

contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour, label='Q(x)')

ax.contour(X, Y, g111_value, [0], colors='red', linewidths=3)
ax.contour(X, Y, g222_value, [0], colors='green', linewidths=3)
ax.contour(X, Y, g333_value, [0], colors='blue', linewidths=3)
ax.contour(X, Y, g444_value, [0], colors='yellow', linewidths=3)

feasible_region = (g111_value >= 0) & (g222_value >= 0) & (g333_value >= 0) & (g444_value >= 0)
ax.contourf(X, Y, feasible_region, levels=[0.5, 1], colors='white', alpha=0.5)

ax.scatter(result.x[0], result.x[1], color='white', label='Оптимальная точка', s=100)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('Рельеф функции и допустимая область')
ax.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.5)

ax.contour(X, Y, g111_value, [0], zdir='z', offset=0, colors='red', linewidths=3)
ax.contour(X, Y, g222_value, [0], zdir='z', offset=0, colors='green', linewidths=3)
ax.contour(X, Y, g333_value, [0], zdir='z', offset=0, colors='blue', linewidths=3)
ax.contour(X, Y, g444_value, [0], zdir='z', offset=0, colors='yellow', linewidths=3)

feasible_region = (g111_value >= 0) & (g222_value >= 0) & (g333_value >= 0) & (g444_value >= 0)
ax.contourf(X, Y, feasible_region, zdir='z', offset=0, levels=[0.5, 1], colors='lightblue', alpha=0.7)

ax.scatter(result.x[0], result.x[1], result.fun, color='black', label='Оптимальная точка')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Q(x)')
ax.set_title('3D визуализация функции и допустимой области')
ax.legend()
plt.show()
```
