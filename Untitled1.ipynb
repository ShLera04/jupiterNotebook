```python
def fib():
    a, b = 0, 1
    while True:
        a, b = b, a + b
        yield a
```

```python
class Patient:
    def __init__(self, name, symptoms, conscious=True, critical=False):
        self.name = name
        self.symptoms = [s.lower() for s in symptoms]  # ["головная боль", "температура"]
        self.conscious = conscious
        self.critical = critical
        self.diagnosis = None
        self.medicine = None

    def is_ready_for_check(self):
        return self.conscious and not self.critical

class SymptomChecker:
    def __init__(self):
        self.rules = {
            ("температура", "кашель", "слабость"): "грипп",
            ("головная боль", "тошнота", "светобоязнь"): "мигрень",
            ("боль в животе", "рвота", "диарея"): "пищевое отравление",
            ("кашель", "одышка", "температура"): "пневмония",
            ("зуд", "высыпания", "отёк"): "аллергия",
        }

    def diagnose(self, patient):
        try:
            if not patient.is_ready_for_check():
                raise RuntimeError("Пациент не в состоянии: нужна экстренная помощь!")
            
            patient_symptoms = tuple(sorted(patient.symptoms))
            
            match patient_symptoms:
                case s if s in self.rules:
                    diag = self.rules[s]
                    print(f"🩺 Диагноз для {patient.name}: {diag}")
                    patient.diagnosis = diag
                case s if any("температура" in sym for sym in s):
                    print(f"🌡️ Подозрение на инфекцию. Нужны анализы.")
                    patient.diagnosis = "инфекция (требуется анализ)"
                case _:
                    print(f"❓ Симптомы не ясны. Назначено наблюдение.")
                    patient.diagnosis = "неопределён"
        except RuntimeError as e:
            print(f"❗ Экстренная ситуация: {e}")
            patient.diagnosis = "требуется реанимация"
        finally:
            print(f"✅ Диагностика для {patient.name} завершена.")

class MedicineStock:
    def __init__(self):
        self.medicines = {
            "грипп": "Тамифлю",
            "мигрень": "Суматриптан",
            "пищевое отравление": "Смекта",
            "пневмония": "Антибиотик (Амоксициллин)",
            "аллергия": "Супрастин",
            "инфекция (требуется анализ)": "Парацетамол"
        }

    def prescribe(self, patient):
        try:
            if patient.diagnosis not in self.medicines:
                raise KeyError(f"Нет лекарства для диагноза: {patient.diagnosis}")
            patient.medicine = self.medicines[patient.diagnosis]
            print(f"💊 Назначено: {patient.medicine}")
        except KeyError as e:
            print(f"⚠️ {e}. Рекомендовано наблюдение.")
            patient.medicine = "нет подходящего лекарства"
        finally:
            print(f"📦 Выписка для {patient.name} сформирована.")

class Clinic:
    def __init__(self):
        self.checker = SymptomChecker()
        self.stock = MedicineStock()

    def process_patient(self, patient):
        print(f"\n--- 🏥 Приём пациента: {patient.name} ---")
        self.checker.diagnose(patient)
        if patient.diagnosis and "требуется реанимация" not in patient.diagnosis:
            self.stock.prescribe(patient)
        print(f"📄 Итог: {patient.name} — {patient.diagnosis}, лечение: {patient.medicine}")
```

```python
def patient_generator():
    names = ["Анна", "Борис", "Вера", "Гриша", "Даша"]
    symptom_sets = [
        ["температура", "кашель", "слабость"],
        ["головная боль", "тошнота"],
        ["боль в животе", "рвота", "диарея"],
        ["кашель", "одышка"],
        ["зуд", "высыпания"],
        ["температура", "головная боль"],
        ["головокружение"],
        ["температура", "кашель", "одышка"]
    ]
    states = [(True, False), (True, False), (False, True), (True, False)]
    i = 0
    while True:
        symptoms = symptom_sets[i % len(symptom_sets)]
        conscious, critical = states[i % len(states)]
        yield Patient(names[i % len(names)], symptoms, conscious, critical)
        i += 1
```

```python
clinic = Clinic()
gen = patient_generator()

for _ in range(5):
    patient = next(gen)
    clinic.process_patient(patient)
```
